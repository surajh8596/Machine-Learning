### HoldOut Validation(Train Test Split)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=40)
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
result=dt.score(x_test,y_test)
print(result*100)
```

### K Fold Cross Validation

```python
from sklearn.model_selection import KFold, cross_val_score
dt=DecisionTreeClassifier()
kf=KFold(n_splits=10)
result=cross_val_score(dt,X,y,cv=kf)
print(result)
print(np.mean(result)*100)
```

### Stratified K Fold Cross Validation

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
dt=DecisionTreeClassifier()
stkf=StratifiedKFold(n_splits=20)
result=cross_val_score(dt,X,y,cv=stkf)
print(result)
print(np.mean(result)*100)
```

### Leave One Out Cross Validation

```python
from sklearn.model_selection import LeaveOneOut, cross_val_score
dt=DecisionTreeClassifier()
leaveone=LeaveOneOut()
result=cross_val_score(dt,X,y,cv=leaveone)
print(result)
print(np.mean(result)*100)
```

### Shuffle Split Cross Validation

```python
from sklearn.model_selection import ShuffleSplit, cross_val_score
dt=DecisionTreeClassifier()
shfl=ShuffleSplit(n_splits=10,test_size=0.30)
result=cross_val_score(dt,X,y,cv=shfl)
print(result)
print(np.mean(result))
```
